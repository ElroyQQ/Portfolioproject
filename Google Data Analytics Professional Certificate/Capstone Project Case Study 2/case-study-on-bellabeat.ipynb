### 2.2 Data bias and credibility:<a class="anchor" id="data_credibility_and_integrity_2_2"></a>

According to the data source on Zenodo(a general-purpose open-access repository), variation between output represents use of different types of Fitbit trackers and individual tracking behaviors / preferences. This might pose a problem of inconsistency in data collection, even though Fitbit is a reputable activity tracker brand.

Consent was given as the data was collected via surveys. Privacy is not a concern as identifiers are not present.

### 3.1 Installing and loading useful packages <a class="anchor" id="installing_loading_packages_3_1"></a>

```r
install.packages("tidyverse") #install the useful collection of R packages for data analysis 
install.packages("janitor") #package useful for data cleaning
install.packages("ggpubr") #package with functions to help customise ggplot
install.packages("sqldf") #package for running SQL statements on R data frames
library(tidyverse) #load the installed packages
library(janitor)
library(ggpubr)
library(sqldf)
```

### 3.2 Importing CSV files <a class="anchor" id="importing_csv_3_2"></a>
 
The 18 CSV files have overlapping data and not all of it may be useful. We will upload the daily data and leave the hourly and minute-level data for later if there is a need. 

After looking through the CSV files in spreadsheets and the metadata, we can determine that the dailyActivity file is the most useful for daily data as it compiles the data for calories, steps and activity intensity.
The sleepDay file contains data on time asleep, time in bed and number of sleep periods exceeding 60 minutes that day. 
The weightLogInfo file contains data on the users' weight.

```r
activity <- read_csv("../input/fitbit/Fitabase Data 4.12.16-5.12.16/dailyActivity_merged.csv") #import the data into data frames
sleep <- read_csv("../input/fitbit/Fitabase Data 4.12.16-5.12.16/sleepDay_merged.csv")
weight <- read_csv("../input/fitbit/Fitabase Data 4.12.16-5.12.16/weightLogInfo_merged.csv")
```

 We will preview our selected data frames and check the summary of each column. 

```r
glimpse(activity) #gives a short glimpse of how the data frames look, including the character types of each column
glimpse(sleep)
glimpse(weight)
```

To be sure the dailyActivity file contains all the data from the individual calories, steps and activity intensity files, we will make use of the sqldf package functions.

```r
calories <- read_csv("../input/fitbit/Fitabase Data 4.12.16-5.12.16/dailyCalories_merged.csv") #import the individual files for checking
steps <- read_csv("../input/fitbit/Fitabase Data 4.12.16-5.12.16/dailySteps_merged.csv") 
intensities <- read_csv("../input/fitbit/Fitabase Data 4.12.16-5.12.16/dailyIntensities_merged.csv")

glimpse(calories)
glimpse(steps)
glimpse(intensities)

#Since activity has 940 rows, we will check if there is any changes in the number of rows if we create the intersection of the data-sets of the two SELECT statements
activity_c <- activity %>% 
    select(Id, ActivityDate, Calories) 
#checking for calories, INNER JOIN instead of INTERSECT as some column names are different
sql_check_c <- sqldf('SELECT * FROM activity_c AS a  
                INNER JOIN calories AS c 
                ON a.Id = c.Id
                AND a.ActivityDate = c.ActivityDay
                AND a.Calories = c.Calories')
glimpse(sql_check_c)

#checking for steps
activity_s <- activity %>% 
    select(Id,ActivityDate, TotalSteps) 
sql_check_s <- sqldf('SELECT * FROM activity_s AS a 
                INNER JOIN steps AS s 
                ON a.Id = s.Id
                AND a.ActivityDate = s.ActivityDay
                AND a.TotalSteps = s.StepTotal')
glimpse(sql_check_s)

#checking for intensities
activity_i <- activity %>% 
    select(Id,ActivityDate, SedentaryMinutes, LightlyActiveMinutes, FairlyActiveMinutes, VeryActiveMinutes, SedentaryActiveDistance, LightActiveDistance, ModeratelyActiveDistance, VeryActiveDistance) 
sql_check_i <- sqldf('SELECT * FROM activity_i AS a 
                INNER JOIN intensities AS i 
                ON a.Id = i.Id
                AND a.ActivityDate = i.ActivityDay
                AND a.SedentaryMinutes = i.SedentaryMinutes
                AND a.LightlyActiveMinutes = i.LightlyActiveMinutes
                AND a.FairlyActiveMinutes = i.FairlyActiveMinutes
                AND a.VeryActiveMinutes = i.VeryActiveMinutes
                AND a.SedentaryActiveDistance = i.SedentaryActiveDistance
                AND a.LightActiveDistance = i.LightActiveDistance
                AND a.ModeratelyActiveDistance = i.ModeratelyActiveDistance
                AND a.VeryActiveDistance = i.VeryActiveDistance')
glimpse(sql_check_i)
```

We can see that the number of rows in each new data frame still remain at 940 even with INNER JOIN, and can conclude that activity contains all the data from the calories, steps and intensities data frame.

### 3.4 Cleaning and formatting <a class="anchor" id="cleaning_and_formatting_3_4"></a>

Before we proceed to do any analysis, we need to clean the data.

From a glance, there are different values in the "Id" field for each data frame. We would want to know how many users contributed to each data frame before deciding if it is useful. 

```r
n_distinct(activity$Id) #count the number of distinct Ids 
n_distinct(sleep$Id)
n_distinct(weight$Id)
```

The activity data frame has 33 unique Ids while sleep has 24. This is interesting as the dataset is supposed to be from 30 fitbit users, a possibility is that the same user uses multiple Fitbit devices. We are unable to identify and remove these without access to other user identifiers, so we will accept this a potential source of bias in our analysis.

For the weight file, it only contains data from 8 users. This is too limited a range for meaningful analysis especially considering the previous point, we will not be using it from now on.

For activity and sleep, we will check for duplicates and remove empty columns or rows.

```r
sum(duplicated(activity)) #checking for duplicate rows in the data frames
sum(duplicated(sleep))
```

There are 3 duplicated rows in sleep, so we will remove them.

```r
sleep <- sleep %>%
  distinct() %>% #only choosing the unique rows and removing duplicates
  drop_na() #remove rows containing missing values
sum(duplicated(sleep)) #checking again for duplicates
```

We also notice that the names of the columns for both data frames are not following the popular naming convention of snake_case. This is an optional step but we will change them just for consistency.

```r
activity <- clean_names(activity) #cleans column names to follow snake_case by default
sleep <- clean_names(sleep)
glimpse(activity) #check to see the changes to the cleaned column names
glimpse(sleep)
```

As observed previously in glimpse(), the date columns for the data frames are in inconsistent or wrong types,we will need to change that. 

We will change the date-time column in "sleep" data frames to only date as we want to combine it with the "activity" data frame for our analysis. After looking through the original csv file in spreadsheets, the times recorded are unnecessary values of either "12:00:00 AM" or "0:00", which are most likely the scheduled time when the data from the smart device is being recorded or synced with the fitbit app.
We also already know the format of the date columns from the metadata file mentioned previously.

```r
activity_clean <- activity %>% 
  rename(date = activity_date) %>% #rename date for consistency
  mutate(date = as.Date(date, "%m/%d/%Y")) #convert to the correct data type
sleep_clean <- sleep %>% 
  rename(date = sleep_day) %>% 
  mutate(date = as.Date(date, format = "%m/%d/%Y %I:%M:%S %p", tz=Sys.timezone())) #convert from date-time to date data type

glimpse(activity_clean) #check for the changes
glimpse(sleep_clean)
```

### 3.5 Merging Data <a class="anchor" id="merging_data_3_5"></a>
Both datas frames now have the 'Id' and 'date' field - this can be used to merge them.
We will combine the data frames by FULL OUTER JOIN, as we may be still able to use the data of users having null values in some variables.


```r
activity_sleep <- merge(activity_clean, sleep_clean, by = c("id","date"), all = TRUE) #merge datasets on the Id and date columns
 glimpse(activity_sleep) #check for the changes we made
 n_distinct(activity_sleep$id) #count the number of unique Ids in our consolidated data frame
 glimpse(activity_sleep)
```

### 3.6 Adding new variables <a class="anchor" id="adding_variables_3_6"></a>
Using the categorisation of physical activity level by steps from WHO(World Health Organisation) publication in 2008 <https://www.who.int/dietphysicalactivity/publications/pacific_pa_guidelines.pdf>, we can categorise our users by level of physical activity as shown below.

Step count | Physical activity level
-----------|------------------------
Less than 5000| Sedentary lifestyle
5000-7499  | Low active
7500-9999  | Somewhat active
Above 10 000| Active
Above 12 500| Highly active

For sleep, we will use the recommendation from the study used by CDC(Centers for Disease Control and Prevention) <https://www.cdc.gov/sleep/about_sleep/how_much_sleep.html> 
* Adults should sleep 7 or more hours per night on a regular basis to promote optimal health 
* Sleeping less than 7 hours per night on a regular basis as too little
* Sleeping more than 9 hours per night on a regular basis as too much"

(We could also incorporate the above information including the adverse/beneficial health outcomes into the app as well)

We would also need to exclude those who did not submit their sleep data.

```r
activity_sleep1 <- activity_sleep %>% #create new data frame to avoid confusion
  mutate(steps_category = case_when( #create a new column with our step categories
    .$total_steps < 5000 ~ "Sedentary",
    .$total_steps >= 5000 &.$total_steps < 7500 ~ "Low_active",
    .$total_steps >= 7500 & .$total_steps < 10000 ~ "Somewhat_Active",
    .$total_steps >= 10000 & .$total_steps < 12500 ~ "Active",
    .$total_steps >= 12500 ~ "Very Active"
  ),sleep_quality = case_when( #create a column for sleep quality
    .$total_minutes_asleep >= 420 &.$total_minutes_asleep <= 540 ~ "Sufficient",
    .$total_minutes_asleep < 420 ~ "Insufficient",
    .$total_minutes_asleep > 540 ~ "Oversleeping"
  ))
glimpse(activity_sleep1)
```

We would also want to explore whether the day of the week plays a part in the level of activity for our data.

```r
wkdy_activity_sleep <- activity_sleep1 %>% #create a new column for days of the week
  mutate(week_day = weekdays(date))

wkdy_activity_sleep$week_day <- factor(wkdy_activity_sleep$week_day, 
                                       c("Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday")) 
#change the data type and put the days in correct order for future plots to be neater
glimpse(wkdy_activity_sleep)
```

Now that we have cleaned the data and added useful variables, we will move on to the analysis.

### 4.2 Activity level analysis <a class="anchor" id="activity_analysis_4_2"></a>


```r
ggplot(activity_sleep1) + 
geom_bar(aes(x = steps_category, fill = steps_category)) + #plot a bar chart of activity level and days recorded
  labs(title = "Total days counted for each activity level", x="Activity level", y="Total number of days", fill= "Activity level")
```

### 4.4 Sleep analysis <a class="anchor" id="sleep_analysis_4_4"></a>

```r
activity_sleep1 %>% 
  drop_na(sleep_quality) %>% #remove null values for the users with no sleep data
  ggplot() + geom_bar(aes(x = sleep_quality, fill = sleep_quality)) + #plot a bar chart of sleep quality and days recorded
  labs(title = "Total days counted for each level of sleep quality", x="Sleep quality", y="Total number of days", fill= "Sleep quality")
```

For sleep quality, we can observe that there is much less days recorded of sleep data, and that oversleeping has a much lower occurence as compared to insufficient or sufficient sleep. We could use this information to drive a campaign with incentives to record their sleep patterns for more data collection as well as bring attention to the benefits of improving sleep quality.

### 4.5 Sleep analysis by day of the week <a class="anchor" id="sleep_analysis_week_4_5"></a>

```r
wkdy_activity_sleep %>% 
  drop_na(sleep_quality) %>% #remove null values for users with no data
  ggplot() + geom_bar(aes(x=sleep_quality,fill=sleep_quality)) + #plot sleep quality vs number of days 
  facet_wrap(~week_day) + #split into individual plots for each day
  labs(title = "Sleep quality for each day of the week", x="Sleep quality", y="Total number of days counted", fill="Sleep quality") +
#add titles to plot for better understanding
   theme(axis.text.x=element_blank(),axis.ticks.x=element_blank()) #remove unnecessary clutter from x axis
```

We can see that oversleeping has the highest occurrence on Sunday, and also higher on Saturday as compared to the weekdays. In combination with the previous observation, we can surmise that the lower number of steps on Sunday might be due to users sleeping in instead of going out to walk or exercise.

### 4.6 Time asleep vs Time in bed<a class="anchor" id="time_asleep_time_bed_4_6"></a>
Next we will look at the relationship between minutes asleep and time in bed. Logically, we would expect it to be almost completely linear.

```r
ggplot(activity_sleep1,aes(x=total_minutes_asleep, y=total_time_in_bed)) + 
  geom_point() + 
  geom_smooth(method="loess") + #Loess Regression to smoothen the volatile time series. 
  labs(title = "Total minutes asleep vs Total time in bed", x="Total minutes asleep", y="Total time in bed") +
  stat_cor(method = "pearson", label.x = 600, label.y = 300) #add correlation coefficient and p-value to an empty space on the plot
```

As predicted, there is a strong positive linear relationship between time in bed and time asleep (correlation coefficient is 0.93), however it is not perfect (a perfect positive correlation would be +1 ) and users spend some time awake in bed. The small p value (2.2x10^-16) that is <0.05 is an indicator of the relationship being statistically significant. 

According to a publication by the National Heart, Lung, and Blood Institute (NHLBI) <https://www.nhlbi.nih.gov/health-topics/all-publications-and-resources/your-guide-healthy-sleep>, it is recommended to not lie in bed awake more than 20 minutes without sleeping, and to get up and do some relaxing activity until you feel sleepy. The anxiety of not being able to sleep can make it harder to fall asleep.
This could be a marketable feature added to the Bellabeat fitness trackers, such as a reminder to the user when the user has spent too much time in bed not being asleep.


### 4.7 Time asleep vs Total distance tracked<a class="anchor" id="time_asleep_total_distance_4_7"></a>

```r
ggplot(wkdy_activity_sleep, aes(x=total_minutes_asleep, y=total_distance)) +
  geom_point() +
  geom_smooth(method="loess") +
  labs(title = "Total minutes asleep vs total distance tracked", x="Total minutes asleep", y="Total distance") +
  stat_cor(method = "pearson", label.x = 200, label.y = 20)
```

There is a weak negative correlation between time asleep and total distance tracked, which would be somewhat expected as people would be more likely to be active if they spend less time asleep.

### 4.8 Calories expended vs Steps taken<a class="anchor" id="calories_steps_4_8"></a>

```r
ggplot(wkdy_activity_sleep, aes(x=calories, y=total_steps)) +
  geom_point() +
  geom_smooth(method="loess") +
  labs(title = "Calories vs Steps taken each day", x="Calories expended", y="Total number of steps") +
  stat_cor(method = "pearson", label.x = 1000, label.y = 30000)
```

Monday and Sunday has the lowest calories expended, while Tuesday has the highest. This is in line with our previous observations for lowest and highest total steps each day.
